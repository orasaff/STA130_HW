```python
# 1. Load the dataset

import pandas as pd
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
df
# df.isna().sum()
```

```python
# 2. Get the number of rows and columns
rows, columns = df.shape
print(f"Rows: {rows}, Columns: {columns}")
```

```python
# Get summary statistics for numerical columns
summary_stats = df.describe()

# Apply basic styling to make it look nicer
styled_summary_stats = summary_stats.style.background_gradient(cmap="Blues").set_caption("Summary Statistics for Titanic Dataset")

# Display the styled DataFrame
styled_summary_stats
```

```python
# Count the occurrences of each value in the 'Pclass' column
class_counts = df['Pclass'].value_counts()

# Convert to DataFrame for styling
class_counts_df = class_counts.reset_index()
class_counts_df.columns = ['Pclass', 'Count']

# Apply basic styling
styled_class_counts = class_counts_df.style.background_gradient(cmap="Greens").set_caption("Passenger Class Counts")

# Display the styled DataFrame
styled_class_counts
```
