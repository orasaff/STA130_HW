# 6. Definition of Summary Statistics

'count' is the sum of how many datapoints exist in this dataset

'mean' is the average of the data values

'std' is short for standard deviation and it is used to shows how much datas are spread out from the mean.

'min' is the smallest value in among the data

'25%' is the distinction of lowest 25% set of data

'50%' is the 50% of data that lies in the middle

'75%' is the distinguish of lowest 75% from the highest 25% dataset

'max' is the largest value in among the data
