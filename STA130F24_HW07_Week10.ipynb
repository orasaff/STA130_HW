But are you sure you know what this means in the context of the **linear form** of a **multiple linear regression** model?_

> - _Imagine you're selling smoothies, the taste of your smoothie depends on the ingredients included in the smoothie, and there are two ingredients (bananas and strawberries) in the smoothie._
>     - _Adding more bananas into the smoothie will of course increase the "banana" flavor of the smoothie, and vice-versa for strawberries..._

>     - _If the "banana" flavor influence is constant and does not depend on the  absolute amount of strawberries in the smoothie, then there is no **interaction** and the **linear form** of the model is $\beta_b b_i + \beta_s s_i$ and the model is said to be only **additive**._

````python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Placeholder dataset (replace this with your actual dataset)
# Assuming 'pokeaman_train' is your DataFrame and 'HP' is the outcome variable

# Set up list to store the in-sample and out-of-sample R-squared values
in_sample_r2 = []
out_of_sample_r2 = []

# Number of iterations for random splits
num_iterations = 100

# For loop to perform the task
for i in range(num_iterations):
    # Split the data randomly into training and testing sets
    train, test = train_test_split(pokeaman_train, test_size=0.2)
    
    # Ensure both training and test sets have the same categories for 'Type 1' and 'Type 2'
    type1_categories = pd.concat([train['Type 1'], test['Type 1']]).unique()
    type2_categories = pd.concat([train['Type 2'], test['Type 2']]).unique()

    # Convert 'Type 1' and 'Type 2' columns to categorical with consistent categories across train and test
    train['Type 1'] = pd.Categorical(train['Type 1'], categories=type1_categories)
    train['Type 2'] = pd.Categorical(train['Type 2'], categories=type2_categories)
    test['Type 1'] = pd.Categorical(test['Type 1'], categories=type1_categories)
    test['Type 2'] = pd.Categorical(test['Type 2'], categories=type2_categories)
    
    # Fit a model using the training data
    model5_linear_form = 'HP ~ Attack + Defense + Speed + Legendary'
    model5_linear_form += ' + Q("Sp. Def") + Q("Sp. Atk")'
    model5_linear_form += ' + C(Generation) + C(Q("Type 1")) + C(Q("Type 2"))'
    
    model5_spec = smf.ols(formula=model5_linear_form, data=train)
    model5_fit = model5_spec.fit()

    # In-sample R-squared (R^2 value for the training data)
    in_sample_r2.append(model5_fit.rsquared)
    
    # Out-of-sample R-squared (R^2 value for the testing data)
    yhat_model5 = model5_fit.predict(test)
    out_of_sample_r2.append(r2_score(test['HP'], yhat_model5))

# Convert the lists to a DataFrame for better visualization
performance_metrics = pd.DataFrame({
    'In-Sample R-squared': in_sample_r2,
    'Out-of-Sample R-squared': out_of_sample_r2
})

# Visualizing the results
plt.figure(figsize=(10, 6))
plt.scatter(performance_metrics['In-Sample R-squared'], performance_metrics['Out-of-Sample R-squared'], alpha=0.5)
plt.title('In-sample vs Out-of-sample R-squared for 100 Random Splits')
plt.xlabel('In-sample R-squared')
plt.ylabel('Out-of-sample R-squared')
plt.grid(True)
plt.show()

# Optionally, print summary statistics
print(performance_metrics.describe())
````
